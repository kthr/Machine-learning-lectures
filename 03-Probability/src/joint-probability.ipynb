```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y2=np.random.normal(0.6,0.2,[30])
y1=np.random.normal(0.4,0.2,[25])
```

```python
y2=np.random.uniform(1,4,[30])
y1=np.random.uniform(2,6,[25])
```

```python
d = {   'y': np.hstack([2 * np.ones([30], dtype=np.int16), np.ones([25],dtype=np.int16)]),
        'x': np.hstack([y2,y1])}
```

```python
df=pd.DataFrame(d)
```

```python
sns.jointplot(x='x',y='y',data=df, size=5, ratio=3, stat_func=None)
```

```python
g = sns.JointGrid(x='x', y='y' , data = df)
g = g.plot_joint(sns.swarmplot, orient='h', color=".5")
g = g.plot_marginals(sns.distplot, hist=True,kde=False, norm_hist=False, color=".5")
```
