```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(style="whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("./data/movie.csv")
```

```python
df
```

```python
snsplot = sns.regplot(x="age", y="liked movie", fit_reg=False, data=df,
               y_jitter=.02)
snsplot.set(ylim=(-0.1,1.1))

snsplot.get_figure().savefig("./img/movie-scatter.svg")
```

```python
snsplot = sns.regplot(x="age", y="liked movie", data=df,
               y_jitter=.02, logistic=False)
snsplot.set(ylim=(-0.1,1.1))

snsplot.get_figure().savefig("./img/movie-linear.svg")
```

```python
snsplot = sns.regplot(x="age", y="liked movie", data=df,
               y_jitter=.02, logistic=True)
snsplot.set(ylim=(-0.1,1.1))

snsplot.get_figure().savefig("./img/movie-logistic.svg")
```

```python
x = np.arange(10,51,1)
y = x<20
y.astype(np.int)
```
